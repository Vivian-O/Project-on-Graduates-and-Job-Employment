QUESTION 1: WHAT IS THE TREND IN UNEMPLOYMENT/EMPLOYMENT OVER THE YEARS?

```python
Q1 = Jobs.loc[:,['Year of graduation', 'Unemployed', 'Employed']]
Q1 = Q1.dropna()
Q1
```

```python
grouped_data = Q1.groupby('Year of graduation').sum()
grouped_data
```

```python
grouped_data.plot.line()
```

QUESTION 2: DOES GENDER PLAY A CRUCIAL PART IN EMPLOYMENT OPPORTUNITY?

```python
Q2 = Jobs.loc[:,['What is your gender', 'Unemployed', 'Employed']]
Q2 = Q2.dropna()
Q2
```

Change the column names


```python
new_columns = {'What is your gender':'Gender'}
Q2= Q2.rename(columns=new_columns)
Q2
```

group the data by gender

```python
grouped_data = Q2.groupby('Gender').sum()
grouped_data
```

Plot the graph now Q2

```python
grouped_data.plot.bar()
```
